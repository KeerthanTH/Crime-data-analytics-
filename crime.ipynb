```python
import pandas as pd
import matplotlib.pyplot as plt

crime=pd.read_csv('../Data sets/crimes_india.csv')
```

```python
crime
```

```python
crime.info()
```

```python
top_10_Burglary=crime[crime['Crime Description']=='BURGLARY']['City'].value_counts(ascending=True).head(10)
```

```python
top_10_Burglary.plot(kind='barh')
```

```python
crime['Date Reported']=pd.to_datetime(crime['Date Reported'],dayfirst=True)
```

```python
crime['month']=crime['Date Reported'].dt.month
```

```python
crime['year']=crime['Date Reported'].dt.year
```

```python
monthly_crimes=crime['month'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,5),facecolor='skyblue')
plt.plot(monthly_crimes.index,monthly_crimes.values,marker='o')
plt.grid(True)
plt.xticks(monthly_crimes.index)
plt.xlabel('Month')
plt.ylabel('Number of Crimes Reported')
plt.title('Monthly Crime Trend')
```

```python
crime=crime(~(crime['month'==12] & (crime['year']>=2024)))
```
